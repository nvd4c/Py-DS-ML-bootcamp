# Categorical Data Plots

Bây giờ chúng ta hãy thảo luận về việc sử dụng seaborn để vẽ dữ liệu phân loại(categorical data)! Có một số loại plot chính cho việc này:

* factorplot
* boxplot
* violinplot
* stripplot
* swarmplot
* barplot
* countplot

Chúng ta hãy xem xét các ví dụ của từng loại!

```python
import seaborn as sns
%matplotlib inline
```

```python
tips = sns.load_dataset('tips')
tips.head()
```

## barplot and countplot

Những plot rất giống nhau này cho phép bạn lấy dữ liệu tổng hợp từ một đối tượng categorical feature trong dữ liệu của mình. **barplot** là một biểu đồ chung cho phép bạn tổng hợp dữ liệu phân loại dựa trên một số chức năng, theo mặc định, nghĩa là:

```python
sns.barplot(x='sex',y='total_bill',data=tips)
```

```python
import numpy as np
```

Bạn có thể thay đổi đối tượng ước lượng(estimator) thành hàm của riêng bạn, chuyển đổi một vectơ thành một đại lượng vô hướng:

```python
sns.barplot(x='sex',y='total_bill',data=tips,estimator=np.std)
```

### countplot

Điều này về cơ bản giống như barplot ngoại trừ công cụ ước tính đang đếm số lần xuất hiện một cách rõ ràng. Đó là lý do tại sao chúng ta chỉ truyền giá trị x:

```python
sns.countplot(x='sex',data=tips)
```

## boxplot and violinplot

boxplots và violinplots được sử dụng để hiển thị sự phân bố của dữ liệu phân loại(categorical data). Biểu đồ hình hộp(box plot) (hoặc box-and-whisker plot) cho thấy sự phân bố dữ liệu định lượng theo cách tạo điều kiện thuận lợi cho việc so sánh giữa các biến hoặc giữa các cấp của một biến phân loại. Box hiển thị các phần tư của tập dữ liệu trong khi phần whiskers mở rộng để hiển thị phần còn lại của phân phối, ngoại trừ các điểm được xác định là "ngoại lệ"(“outliers”) bằng cách sử dụng một phương pháp là một hàm của phạm vi giữa các phần tư(function of the inter-quartile range).

```python
sns.boxplot(x="day", y="total_bill", data=tips,palette='rainbow')
```

```python
# Có thể thực hiện toàn bộ khung dữ liệu với orient='h' , định hướng = 'h'
sns.boxplot(data=tips,palette='rainbow',orient='h')
```

```python
sns.boxplot(x="day", y="total_bill", hue="smoker",data=tips, palette="coolwarm")
```

### violinplot

A violin plot có vai trò tương tự như plot của box and whisker plot. Nó cho thấy sự phân bố của dữ liệu định lượng qua một số cấp độ của một (hoặc nhiều) biến phân loại để có thể so sánh những phân phối đó. Không giống như một biểu đồ box, trong đó tất cả các thành phần của biểu đồ tương ứng với các điểm dữ liệu thực tế, violin plot có tính năng ước tính mật độ nhân của phân phối cơ bản.

```python
sns.violinplot(x="day", y="total_bill", data=tips,palette='rainbow')
```

```python
sns.violinplot(x="day", y="total_bill", data=tips,hue='sex',palette='Set1')
```

```python
sns.violinplot(x="day", y="total_bill", data=tips,hue='sex',split=True,palette='Set1')
```

## stripplot and swarmplot

Stripplot sẽ vẽ một biểu đồ phân tán(scatterplot) trong đó một biến được phân loại. Biểu đồ dải(strip plot) có thể được vẽ theo cách riêng của nó, nhưng nó cũng là một bổ sung tốt cho biểu đồ box or violin plot trong trường hợp bạn muốn hiển thị tất cả các quan sát cùng với một số biểu diễn của phân phối cơ bản.

Swarmplot tương tự như stripplot(), nhưng các điểm được điều chỉnh (chỉ dọc theo trục phân loại) để chúng không trùng nhau. Điều này cho thấy sự thể hiện tốt hơn về sự phân bố của các giá trị, mặc dù nó không mở rộng quy mô cũng như đối với số lượng lớn các quan sát (cả về khả năng hiển thị tất cả các điểm và về tính toán cần thiết để sắp xếp chúng).

```python
sns.stripplot(x="day", y="total_bill", data=tips)
```

```python
sns.stripplot(x="day", y="total_bill", data=tips,jitter=True)
```

```python
sns.stripplot(x="day", y="total_bill", data=tips,jitter=True,hue='sex',palette='Set1')
```

```python
sns.stripplot(x="day", y="total_bill", data=tips,jitter=True,hue='sex',palette='Set1',split=True)
```

```python
sns.swarmplot(x="day", y="total_bill", data=tips)
```

```python
sns.swarmplot(x="day", y="total_bill",hue='sex',data=tips, palette="Set1", split=True)
```

### Combining Categorical Plots

```python
sns.violinplot(x="tip", y="day", data=tips,palette='rainbow')
sns.swarmplot(x="tip", y="day", data=tips,color='black',size=3)
```

## factorplot

factorplot là hình thức chung nhất của một categorical plot. Nó có thể sử dụng tham số **kind** để điều chỉnh loại plot:

```python
sns.factorplot(x='sex',y='total_bill',data=tips,kind='bar')
```
