# Regression Plots

Seaborn có nhiều khả năng tích hợp cho các biểu đồ hồi quy, tuy nhiên chúng ta sẽ không thực sự thảo luận về hồi quy cho đến phần học máy của khóa học, vì vậy hiện tại chúng ta sẽ chỉ đề cập đến hàm **lmplot()**.

**lmplot** cho phép bạn hiển thị các mô hình tuyến tính, nhưng nó cũng cho phép bạn tách các ô đó một cách thuận tiện dựa trên các tính năng, cũng như tô màu sắc độ dựa trên các đặc điểm.

Hãy khám phá cách thức hoạt động của điều này:

```python
import seaborn as sns
%matplotlib inline
```

```python
tips = sns.load_dataset('tips')
```

```python
tips.head()
```

## lmplot()

```python
sns.lmplot(x='total_bill',y='tip',data=tips)
```

```python
sns.lmplot(x='total_bill',y='tip',data=tips,hue='sex')
```

```python
sns.lmplot(x='total_bill',y='tip',data=tips,hue='sex',palette='coolwarm')
```

### Working with Markers

lmplot kwargs được chuyển đến **regplot**, một dạng tổng quát hơn của lmplot(). regplot có tham số scatter_kws được chuyển tới plt.scatter. Vì vậy, bạn muốn đặt tham số s trong từ điển đó, tương ứng (hơi khó hiểu) với kích thước điểm đánh dấu bình phương. Nói cách khác, bạn sẽ chuyển một từ điển với các đối số matplotlib cơ sở, trong trường hợp này, là kích thước của một biểu đồ phân tán. Nói chung, bạn có thể sẽ không nhớ điều này ra khỏi đầu, nhưng thay vào đó hãy tham khảo tài liệu.

```python
# http://matplotlib.org/api/markers_api.html
sns.lmplot(x='total_bill',y='tip',data=tips,hue='sex',palette='coolwarm',
           markers=['o','v'],scatter_kws={'s':100})
```

## Using a Grid

Chúng ta có thể thêm nhiều biến phân tách thông qua các cột và hàng bằng cách sử dụng grid. Chỉ cần chỉ ra điều này bằng các đối số col hoặc row:

```python
sns.lmplot(x='total_bill',y='tip',data=tips,col='sex')
```

```python
sns.lmplot(x="total_bill", y="tip", row="sex", col="time",data=tips)
```

```python
sns.lmplot(x='total_bill',y='tip',data=tips,col='day',hue='sex',palette='coolwarm')
```

## Aspect and Size

Seaborn figure có thể được điều chỉnh kích thước và tỷ lệ khung hình với các thông số **size** and **aspect** parameters:

```python
sns.lmplot(x='total_bill',y='tip',data=tips,col='day',hue='sex',palette='coolwarm',
          aspect=0.6,size=8)
```

Có thể bạn đang tự hỏi làm thế nào để thay đổi kích thước phông chữ hoặc kiểm soát tính thẩm mỹ hơn nữa, hãy xem Bài giảng Style and Color và Notebook để biết thêm thông tin về điều đó!
